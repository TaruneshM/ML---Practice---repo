## Importig lib


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing datasets

```python
df=pd.read_csv('Social_Network_Ads.csv')
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

## Spliting the data into test set and training sets

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.25,random_state=0)
```

```python
x_train
```

## Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

## Training the SVM model

```python
from sklearn.svm import SVC
classi=SVC(kernel='linear',random_state=0)
classi.fit(x_train,y_train)
```

## Predicting a new result

```python
print(classi.predict(sc.transform([[30,87000]])))
```

## Predicting the test set result

```python
y_pred=classi.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

## Making the confusion matrix

```python
from sklearn.metrics import confusion_matrix , accuracy_score
cm=confusion_matrix(y_test , y_pred)
print(cm)
accuracy_score(y_test , y_pred)
```

## Visualising the training set result

```python
from matplotlib.colors import ListedColormap
x_set,y_set=sc.inverse_transform(x_train),y_train
x1,x2=np.meshgrid(np.arange(start=x_set[:,0].min()-10,stop=x_set[:,0].max()+10,step=0.25),
                  np.arange(start=x_set[:,1].min()-1000,stop=x_set[:,1].max()+1000 , step=0.25))
plt.contourf(x1,x2,classi.predict(sc.transform(np.array([x1.ravel(),x2.ravel()]).T)).reshape(x1.shape),
             alpha=0.75,cmap=ListedColormap(('red','green')))
plt.xlim(x1.min() , x1.max())
plt.ylim(x2.min(),x2.max())
for i , j in enumerate(np.unique(y_set)):
  plt.scatter(x_set[y_set==j,0],x_set[y_set==j,1],c=ListedColormap(('red','green'))(i),label=j)
plt.title('Logistic reg')
plt.xlabel('age')
plt.ylabel('estimated salary')
plt.legend()
plt.show()
```
